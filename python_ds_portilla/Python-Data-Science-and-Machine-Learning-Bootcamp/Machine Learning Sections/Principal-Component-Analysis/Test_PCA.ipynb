```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
```

```python
df.head()
```

```python
cancer['target']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca
```

```python
plt.figure(figsize = (8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = cancer['target'], cmap = 'plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns = cancer['feature_names'])
```

```python
df_squares = pd.DataFrame(pca.components_**2, columns = cancer['feature_names'])
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(df_squares, cmap = 'plasma')
```

```python
from sklearn.cluster import KMeans
```

```python
reduced = pd.DataFrame(data = x_pca, columns = ['First PC', 'Second PC'])
```

```python
fitter = KMeans(n_clusters=2)
```

```python
fitter.fit(reduced)
```

```python
fitter.labels_
reduced['Cluster'] = fitter.labels_
```

```python
sns.lmplot(x = 'First PC', y = 'Second PC', data = reduced, hue = 'Cluster')
```
