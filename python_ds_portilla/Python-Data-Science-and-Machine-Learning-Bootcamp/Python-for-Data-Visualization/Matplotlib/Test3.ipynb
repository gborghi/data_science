```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
plt.plot(x,y)
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
#axes1.set_aspect('equal')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
for current_ax in axes:
    current_ax.plot(x,y)
axes[0].plot(y,x)
axes[0].set_title("Primo")
axes[1].set_title("Secondo")
```

```python
fig=plt.figure(figsize=(8,8), dpi=250)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig.savefig('mypicture.png', dpi=200)
```

```python
fig=plt.figure(figsize=(4,4))
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label="plotino")
ax.plot(x,x**3, label="paperone")

ax.set_ylabel('Y')

ax.legend(loc=(0.1,0.1))
```
